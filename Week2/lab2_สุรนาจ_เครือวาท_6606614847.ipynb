## 4) เลือกดาวเทียม ช่วงเวลา และเกณฑ์เมฆ

### เหตุผลที่เลือกดาวเทียม: Sentinel-2 SR Harmonized
- ความละเอียดเชิงพื้นที่ 10 เมตร เหมาะกับงานระดับจังหวัด
- มี band ที่เหมาะกับการคำนวณดัชนีด้านพืชพรรณ ความชื้น และแหล่งน้ำ
- เหมาะกับการวิเคราะห์ช่วงฤดูแล้งที่ต้องการแยกความต่างของพื้นที่ให้ชัด

### ช่วงเวลาที่เลือก
- **Period A (ต้นฤดูแล้ง):** 2024-11-01 ถึง 2025-01-31  
- **Period B (ปลายฤดูแล้ง):** 2025-02-01 ถึง 2025-04-30  

### เกณฑ์เมฆ
- ใช้ `CLOUDY_PIXEL_PERCENTAGE < 20`
- ถ้ากรองเมฆเข้มเกินไป จำนวนภาพจะลดลง
- ถ้าผ่อนเกณฑ์มากเกินไป composite จะมี noise จากเมฆหรือเงาเมฆ


```python
START_A = '2024-11-01'
END_A   = '2025-01-31'
START_B = '2025-02-01'
END_B   = '2025-04-30'
CLOUD_FILTER = 20
```

```python
def mask_s2_sr(image):
    scl = image.select('SCL')
    mask = (scl.neq(3)    # cloud shadow
            .And(scl.neq(8))   # medium probability cloud
            .And(scl.neq(9))   # high probability cloud
            .And(scl.neq(10))  # thin cirrus
            .And(scl.neq(11))) # snow / ice
    return image.updateMask(mask).divide(10000)

def add_indices(img):
    ndvi  = img.normalizedDifference(['B8', 'B4']).rename('NDVI')
    ndmi  = img.normalizedDifference(['B8', 'B11']).rename('NDMI')
    mndwi = img.normalizedDifference(['B3', 'B11']).rename('MNDWI')
    return img.addBands([ndvi, ndmi, mndwi])

def make_s2_composite(start_date, end_date, region):
    col = (ee.ImageCollection("COPERNICUS/S2_SR_HARMONIZED")
           .filterBounds(region)
           .filterDate(start_date, end_date)
           .filter(ee.Filter.lt('CLOUDY_PIXEL_PERCENTAGE', CLOUD_FILTER))
           .map(mask_s2_sr))
    comp = add_indices(col.median().clip(region))
    return col, comp

col_a, comp_a = make_s2_composite(START_A, END_A, aoi)
col_b, comp_b = make_s2_composite(START_B, END_B, aoi)

print("Number of images in Period A:", col_a.size().getInfo())
print("Number of images in Period B:", col_b.size().getInfo())
```

## 6) วิเคราะห์ที่ 1 — Spectral Index

ในงานนี้เลือกใช้ 3 ดัชนีหลัก
- **NDVI**: ความเขียวของพืชพรรณ
- **NDMI**: ความชื้นของพืชพรรณและผิวดิน
- **MNDWI**: การเน้นแหล่งน้ำผิวดิน

> ตอนเขียนรายงาน อย่าเขียนแค่ “ค่าสูง/ต่ำ” แต่ต้องอธิบายว่าเห็นตรงไหน และหมายความว่าอะไรในบริบทพื้นที่จริง


```python
vis_rgb   = {'bands': ['B4', 'B3', 'B2'], 'min': 0.0, 'max': 0.3}
vis_ndvi  = {'min': -1, 'max': 1, 'palette': ['#8c510a', '#f6e8c3', '#01665e']}
vis_ndmi  = {'min': -1, 'max': 1, 'palette': ['#d73027', '#fdae61', '#1a9850']}
vis_mndwi = {'min': -1, 'max': 1, 'palette': ['#8c510a', '#f6e8c3', '#2166ac']}

Map = geemap.Map(center=[18.77, 100.77], zoom=8)
Map.addLayer(comp_b, vis_rgb, 'RGB Period B')
Map.addLayer(comp_b.select('NDVI'), vis_ndvi, 'NDVI Period B')
Map.addLayer(comp_b.select('NDMI'), vis_ndmi, 'NDMI Period B')
Map.addLayer(comp_b.select('MNDWI'), vis_mndwi, 'MNDWI Period B')
Map.addLayer(districts_fc.style(**{"color": "black", "fillColor": "00000000"}), {}, "District boundaries")
Map
```

### การตีความดัชนี ได้เเก่ NDVI, NDMI, เเละ MNDWI
- พื้นที่ NDVI สูงต่อเนื่องในเขตภูเขาและป่าไม้ สะท้อนพื้นที่ที่ยังมีพืชพรรณหนาแน่น
- พื้นที่ NDMI สูงชี้ถึงบริเวณที่มีความชื้นสะสมมากกว่า
- พื้นที่ MNDWI สูงมักสัมพันธ์กับลำน้ำ อ่างเก็บน้ำ หรือแหล่งน้ำผิวดิน


## 7) วิเคราะห์ที่ 2 — เปรียบเทียบ Composite 2 ช่วงเวลา
ใช้ค่า `ΔNDVI = NDVI ช่วงปลายฤดูแล้ง - NDVI ช่วงต้นฤดูแล้ง`


```python
delta_ndvi = comp_b.select('NDVI').subtract(comp_a.select('NDVI')).rename('dNDVI')

vis_dndvi = {
    'min': -0.5,
    'max': 0.5,
    'palette': ['#b2182b', '#fddbc7', '#f7f7f7', '#d1e5f0', '#2166ac']
}

Map = geemap.Map(center=[18.77, 100.77], zoom=8)
Map.addLayer(comp_a.select('NDVI'), vis_ndvi, 'NDVI Period A')
Map.addLayer(comp_b.select('NDVI'), vis_ndvi, 'NDVI Period B')
Map.addLayer(delta_ndvi, vis_dndvi, 'Delta NDVI (B - A)')
Map.addLayer(districts_fc.style(**{"color": "black", "fillColor": "00000000"}), {}, "District boundaries")
Map
```

### แนวทางตีความที่ควรเขียน
- ค่า `dNDVI` ติดลบมาก หมายถึงความเขียวลดลง
- ค่าใกล้ศูนย์ หมายถึงการเปลี่ยนแปลงน้อย
- ค่าเป็นบวก อาจสะท้อนพื้นที่ที่ยังมีน้ำหรือมีการเพาะปลูกในช่วงปลายฤดูแล้ง


## 8) วิเคราะห์ที่ 3 — Zonal Statistics ระดับอำเภอ
จะคำนวณค่าเฉลี่ย NDVI, NDMI และ MNDWI แยกตามอำเภอ


```python
# เลือกชื่อคอลัมน์อำเภอจาก shapefile อัตโนมัติ
district_name_candidates = [
    "ADM2_NAME", "NAME_2", "AMP_NAME_T", "AMP_NAM_T", "DISTRICT", "district", "NAME"
]
district_name_field = None
for col in districts_gdf.columns:
    if col in district_name_candidates:
        district_name_field = col
        break

if district_name_field is None:
    print("Please inspect your district shapefile columns and set district_name_field manually.")
else:
    print("District name field:", district_name_field)
```

```python
zonal_stats = comp_b.select(['NDVI', 'NDMI', 'MNDWI']).reduceRegions(
    collection=districts_fc,
    reducer=ee.Reducer.mean(),
    scale=10
)

zonal_df = geemap.ee_to_df(zonal_stats)

# เลือกคอลัมน์ชื่ออำเภอสำหรับแสดงผล
if district_name_field in zonal_df.columns:
    display_name_col = district_name_field
else:
    display_name_col = zonal_df.columns[0]

zonal_df = zonal_df[[display_name_col, 'NDVI', 'NDMI', 'MNDWI']].copy()
zonal_df = zonal_df.sort_values('NDVI', ascending=False).reset_index(drop=True)
zonal_df.head(10)
```

```python
plt.figure(figsize=(10, 6))
plt.barh(zonal_df[display_name_col], zonal_df['NDVI'])
plt.gca().invert_yaxis()
plt.title('Average NDVI by District (Period B)')
plt.xlabel('Mean NDVI')
plt.ylabel('District')
plt.tight_layout()
plt.show()
```

### แนวทางตีความที่ควรเขียน
- อำเภอที่ NDVI สูง อาจมีพื้นที่ป่าหรือพื้นที่พืชพรรณหนาแน่นมากกว่า
- อำเภอที่ NDMI สูง อาจมีความชื้นสะสมมากกว่า
- อำเภอที่ MNDWI สูง อาจมีสัดส่วนแหล่งน้ำผิวดินหรือพื้นที่ชุ่มน้ำมากกว่า


## 9) วิเคราะห์ที่ 4 — ความสัมพันธ์ระหว่าง NDVI และ NDMI

## 10) วิเคราะห์ที่ 5 — คำนวณพื้นที่พืชพรรณหนาแน่นพร้อม uncertainty
ใช้เกณฑ์ 2 ค่าเพื่อสะท้อนความไม่แน่นอนของ threshold

- กรณีที่ 1: `NDVI > 0.50`
- กรณีที่ 2: `NDVI > 0.55`


```python
pixel_area_km2 = ee.Image.pixelArea().divide(1e6).rename('area')

dense_050 = comp_b.select('NDVI').gt(0.50).selfMask()
dense_055 = comp_b.select('NDVI').gt(0.55).selfMask()

area_050_val = pixel_area_km2.updateMask(dense_050).reduceRegion(
    reducer=ee.Reducer.sum(),
    geometry=aoi,
    scale=10,
    maxPixels=1e13
).get('area').getInfo()

area_055_val = pixel_area_km2.updateMask(dense_055).reduceRegion(
    reducer=ee.Reducer.sum(),
    geometry=aoi,
    scale=10,
    maxPixels=1e13
).get('area').getInfo()

print("Dense vegetation area (NDVI > 0.50), sq.km:", area_050_val)
print("Dense vegetation area (NDVI > 0.55), sq.km:", area_055_val)
print("Uncertainty range, sq.km:", abs(area_050_val - area_055_val))
```

## 11) เปรียบเทียบกับดาวเทียมอีกดวง (Landsat 9)
ส่วนนี้เอาไว้ใช้ตอบคำถามท้าย README ว่าถ้าเลือกดาวเทียมอีกดวง ผลจะต่างกันอย่างไร


```python
def mask_landsat_l2(image):
    qa = image.select('QA_PIXEL')
    cloud = qa.bitwiseAnd(1 << 3).eq(0)
    shadow = qa.bitwiseAnd(1 << 4).eq(0)
    snow = qa.bitwiseAnd(1 << 5).eq(0)
    return image.updateMask(cloud.And(shadow).And(snow))

def scale_landsat(img):
    optical = img.select(['SR_B2', 'SR_B3', 'SR_B4', 'SR_B5', 'SR_B6']).multiply(0.0000275).add(-0.2)
    ndvi  = optical.normalizedDifference(['SR_B5', 'SR_B4']).rename('NDVI')
    ndmi  = optical.normalizedDifference(['SR_B5', 'SR_B6']).rename('NDMI')
    mndwi = optical.normalizedDifference(['SR_B3', 'SR_B6']).rename('MNDWI')
    return optical.addBands([ndvi, ndmi, mndwi])

landsat9 = (ee.ImageCollection("LANDSAT/LC09/C02/T1_L2")
            .filterBounds(aoi)
            .filterDate(START_B, END_B)
            .map(mask_landsat_l2))

landsat9_comp = landsat9.map(scale_landsat).median().clip(aoi)

Map = geemap.Map(center=[18.77, 100.77], zoom=8)
Map.addLayer(comp_b.select('NDVI'), vis_ndvi, 'Sentinel-2 NDVI')
Map.addLayer(landsat9_comp.select('NDVI'), vis_ndvi, 'Landsat 9 NDVI')
Map
```

### ถ้าเลือกดาวเทียมอีกดวง ผลจะต่างอย่างไร?

##สรุปสิ่งที่ค้นพบ

จังหวัดน่านยังคงมีพื้นที่พืชพรรณหนาแน่นจำนวนมากในช่วงปลายฤดูแล้ง โดยเฉพาะในอำเภอที่มีองค์ประกอบของภูเขาและป่าสูง ซึ่งสะท้อนผ่านค่า NDVI และ NDMI ที่ค่อนข้างสูงในหลายพื้นที่ ค่าเฉลี่ยระดับอำเภอชี้ชัดว่าอำเภออย่าง อําเภอบ่อเกลือ และ อําเภอสองเเคว มีทั้งความเขียวและความชื้นเฉลี่ยสูงเด่นกว่าหลายอำเภออื่น

ในอีกด้านหนึ่ง การเปรียบเทียบสองช่วงเวลาทำให้เห็นว่างานวิเคราะห์แบบ time composite ไม่ได้บอกแค่ “ค่าสูงหรือต่ำ” แต่ช่วยชี้ให้เห็นพลวัตของฤดูแล้ง ว่าพื้นที่ใดลดความเขียวลง พื้นที่ใดคงตัว หรือพื้นที่ใดอาจยังรักษาความชื้นและพืชพรรณไว้ได้ ซึ่งเป็นสารสนเทศที่มีคุณค่าในเชิงการจัดการพื้นที่

ผลสหสัมพันธ์ที่สูงมากระหว่าง NDVI กับ NDMI ยังสนับสนุนแนวคิดว่าพื้นที่ที่ยังเขียวมากในจังหวัดน่านมักเป็นพื้นที่ที่ยังมีความชื้นสะสมอยู่พอสมควรด้วย และผลการคำนวณพื้นที่พืชพรรณหนาแน่นพร้อม uncertainty ก็ช่วยตอกย้ำว่าการรายงานผลเชิงพื้นที่ควรระบุเงื่อนไขและสมมติฐานไว้ชัดเจน ไม่ควรนำเสนอเพียงค่าตัวเลขเดียวโดยไม่มีบริบท

กล่าวโดยสรุป Lab 2 spatial_analysis บรรลุเป้าหมายทั้งการเลือกพื้นที่และเหตุผล การเลือกดาวเทียมและช่วงเวลา การทำงานวิเคราะห์มากกว่า 3 ข้อ การแปลความหมายเชิงพื้นที่ การทดลองเปรียบเทียบดาวเทียมอีกดวง และการ export ผลลัพธ์ในรูปแบบที่ใช้งานต่อได้จริง
